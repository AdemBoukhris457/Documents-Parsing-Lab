## **2. Parse pdf files with Zerox**

```python
def display_pdf_page(pdf_path_to_display, page_numbers=None):
    """
    Display selected pages of a PDF as images.

    Args:
        pdf_path_to_display (str): Path to the PDF file.
        page_numbers (list, optional): List of page numbers to display (1-indexed). 
                                        If None, display all pages.
    """
    try:
        # Convert the PDF to images
        images = convert_from_path(pdf_path_to_display)
        
        # If specific pages are requested, filter the images
        if page_numbers:
            images = [images[i - 1] for i in page_numbers if 1 <= i <= len(images)]

        # Display each selected page
        for i, image in enumerate(images):
            plt.figure(figsize=(16, 12))
            plt.imshow(image)
            plt.axis('off')  # Hide axis for a cleaner look
            plt.title(f"Page {i + 1}")
            plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
# Define main async entrypoint
async def parse_pdf_file(file_path):
    kwargs = {}
    custom_system_prompt = None
    model = "gemini/gemini-1.5-pro-002"
    select_pages = None 
    output_dir = "parsed_data" ## directory to save the consolidated markdown file
    result = await zerox(file_path=file_path, model=model, output_dir=output_dir,
                        custom_system_prompt=custom_system_prompt,select_pages=select_pages, **kwargs)
    return result
```

```python
display_pdf_page("pdf_files/NVIDIA_Report_Removed_Pages-1-16-1.pdf")
result = asyncio.run(parse_pdf_file("pdf_files/NVIDIA_Report_Removed_Pages-1-16-1.pdf"))
print(result.pages[0].content)
#display(Markdown(result.pages[0].content))
```
